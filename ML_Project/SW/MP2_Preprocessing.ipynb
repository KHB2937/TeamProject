```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

Jan_data = pd.read_csv('https://github.com/jaiwon880/ML_Projiect/raw/main/Data/%EC%95%84%ED%8C%8C%ED%8A%B8_%EB%A7%A4%EB%A7%A4__%EC%8B%A4%EA%B1%B0%EB%9E%98%EA%B0%80_20230321134107_1%EC%9B%94.csv')
Feb_Data = pd.read_csv('https://raw.githubusercontent.com/jaiwon880/ML_Projiect/main/Data/%EC%95%84%ED%8C%8C%ED%8A%B8_%EB%A7%A4%EB%A7%A4__%EC%8B%A4%EA%B1%B0%EB%9E%98%EA%B0%80_20230321134120_2%EC%9B%94.csv')

def Pre(datas):
    data=[]
    # datas['거래금액(만원)'].str.strip().apply(lambda x: x.replace(',',''))
    
    datas['거래금액(만원)']=(datas['거래금액(만원)'].str.replace(',','')).str.strip().astype(int)
    datas["평당가"] = (datas["거래금액(만원)"] / datas["전용면적(㎡)"]*0.3025) # 평당가 계산하여 새로운 컬럼 추가
    aa=datas['계약년월'].astype(str)
    bb=datas['계약일'].astype(str)
    datas['date'] = aa+bb
    datas['date'] = pd.to_datetime(datas['date'], format='%Y%m%d')
    datas['금리'] = datas.apply(lambda x: 3.25 if ((x['계약년월']== 202301) & (x['계약일'] < 13)) else 3.5, axis=1)
    datas.drop(['계약년월','계약일'],axis=1,inplace=True)
    datas['거래유형'] = datas['거래유형'].apply(lambda x:0 if x=='중개거래' else 1)
    datas.dropna(inplace=True)
    data= datas[datas['시군구']=='충청남도 계룡시 금암동']    

    return data
    
df1= Pre(Jan_data)
```
