```python
def preprocessing(data):
    datas=[]
    data['거래금액(만원)']=(data['거래금액(만원)'].str.replace(',','')).str.strip().astype(int)
    data["평당가"] = (data["거래금액(만원)"] / data["전용면적(㎡)"]*0.3025)
    aa=data['계약년월'].astype(str)
    bb=data['계약일'].astype(str)
    data['계약날짜'] = aa+bb
    data['계약날짜'] = pd.to_datetime(data['계약날짜'], format='%Y%m%d')
    
    data['거래유형'] = data['거래유형'].apply(lambda x:0 if x=='중개거래' else 1)
    data["금리"]=data.apply(lambda x: 3.25 if ((x['계약년월']== 202301) & (x['계약일'] < 13)) else 3.5, axis=1)
    data.dropna(inplace=True)
    data.drop(['계약년월','계약일'],axis=1,inplace=True)
    datas=data.copy()
    return datas
```

```python
preprocessing(df1)
```

```python
def Modeling(data):
  count=0
  min_value = float(8000.2)
  X=data.drop('거래금액(만원)',axis=1)
  y=data["거래금액(만원)"]
  for i in range(1000,1050):
    X_train, X_test, y_train, y_test = train_test_split(
          X, y, test_size = 0.2, random_state=i
    )
    model = DecisionTreeClassifier(random_state=i)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = mean_squared_error(y_test,y_pred)**0.5
    print(rmse, i)
    if rmse < min_value:
        min_value = rmse
        count=i
  print(min_value, count)
```

```python
Modeling(df)
```
